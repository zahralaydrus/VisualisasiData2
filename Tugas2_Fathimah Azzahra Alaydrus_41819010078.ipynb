```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
```

```python
np.random.seed(10)
x= np.random.rand(100)
y= np.random.randint(10, 70, 100)
madrid =['isco', 'benzema', 'ronaldo']
pemain = np.random.choice(madrid, 100)
```

```python
zahra = pd.DataFrame(dict(X=x, Y=y, pemain=pemain))
zahra.head(10)
```

```python
colors = {'isco':'red', 'benzema':'green', 'ronaldo':'black'}

fig, ax = plt.subplots()
ax.scatter(zahra['X'], zahra['Y'], c=zahra['pemain'].map(colors))
plt.show()
```

```python
sns.scatterplot('X', 'Y', data=zahra, hue='pemain')
```

```python
fig, ax = plt.subplots()

zahra_group = zahra.groupby('pemain')
colors = {'isco':'red', 'benzema':'green', 'ronaldo':'black'}

for name, group in zahra_group:
    group.plot(ax=ax, kind='scatter', x='X', y='Y', label=name, color=colors[name])

plt.show()
```

```python
y=np.array([40,30,30])
myLabels=["isco", "benzema", "ronaldo"]
mycolors=["red", "green", "black"]


plt.pie(y, labels=myLabels, colors=mycolors)
plt.legend( title="pemain")
plt.show

sys.stdout.flush()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
pemain=['isco', 'benzema', 'ronaldo']
Nilai=[100,80,100]
ax.bar(pemain, Nilai, color='blue')
plt.legend(title="pemain")
plt.show()
```
